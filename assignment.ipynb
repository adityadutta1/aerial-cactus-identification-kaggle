PROJECT DESCRIPTION:
Identifying specific type of cactus in aerial imagery.
keras and convolution neural network are used to solve the problem. 
Problem Link is https://www.kaggle.com/c/aerial-cactus-identification

Importing the necessary packages,splitting the dataset into test and train 

```python
import numpy as np 
import pandas as pd 
import tensorflow as tf 
tf.keras.models.Sequential
tf.keras.layers.Dense
tf.keras.layers.Conv2D 
tf.keras.layers.Flatten
tf.keras.layers.Dropout
import cv2
from sklearn.model_selection import train_test_split
import os
import random
import matplotlib.pyplot as plt
import csv

print(os.listdir("C:/Users/Adity/Desktop/projo"))
train_csv = pd.read_csv('C:/Users/Adity/Desktop/projo/train.csv').sample(frac=1).reset_index(drop=True)
images = train_csv.id.values.tolist() 
target = train_csv.has_cactus.values.tolist()
train_X, test_X, train_Y, test_Y = train_test_split(images, target, test_size=0.1, random_state=42)
del train_csv, images, target
```

Reading the images in the train data using open cv library.

```python
def get_image(imname):
    name = os.path.join('C:/Users/Adity/Desktop/projo/train', imname)
    img = cv2.imread(name, 1)
    img = cv2.resize(img, (32,32))/255
    return img
```


Preprocessing the the train and test images before applying convolution neural networks
Reshaping the numpy arrays.

```python
batch_img = []
batch_tar = []
val_x = []
val_y = []
for i in range(len(train_X)):
    batch_img.append(np.reshape(get_image(train_X[i]), (32,32,3)))
    batch_tar.append(train_Y[i])
for i in range(len(test_X)):
    val_x.append(np.reshape(get_image(test_X[i]), (32,32,3)))
    val_y.append(test_Y[i])
batch_img, val_x = np.array(batch_img), np.array(val_x)
```


MODEL:

The Sequential model is a linear stack of layers.
We can create a Sequential model by passing a list of layer instances to the constructor.
We can also simply add layers via the .add() method.
we will be using 32 by 32 RGB pictures in our project using input_shape argument.
kernel size specifying the height and width of the 2D convolution window is 3 by 3
first argument in Conv2D layers signifies the number of output filters in the convolution.
Padding as "same" results in padding the input such that the output has the same length as the original input.
One way ReLUs improve neural networks is by speeding up training.Therefore we use RELU as an activation function.
Then we flatten the output and feed into a fully connected layer.
Here we use Dense layer or Fully connected in which every node in the layer is connected to every node in the preceding layer.
We use Sigmoid as an activation Function to find how likely it is that the image falls into each target class.
A metric is a function that is used to judge the performance of your model. Metric functions are to be supplied in the metrics parameter when a model is compiled.This is our last step.


```python
model = Sequential()
model.add(Conv2D(16, kernel_size=3, padding='same', activation='relu', input_shape=(32,32,3)))
model.add(Conv2D(16, kernel_size=3, padding='same', activation='relu'))
model.add(Conv2D(8, kernel_size=3, padding='same', activation='relu'))
model.add(Flatten())
model.add(Dense(1, activation='sigmoid'))

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

Keras models are trained on Numpy arrays of input data and labels. For training a model, we will typically use the fit function.
Achieved accuracy of 99%

```python
model.fit(batch_img, batch_tar, validation_data = (val_x, val_y), epochs = 20)
```

Reading the images in test data with help of imread function of cv2.

```python
test_list = os.listdir('C:/Users/Adity/Desktop/projo/test')
def get_test_image(imname):
    name = os.path.join('C:/Users/Adity/Desktop/projo/test', imname)
    img = cv2.imread(name, 1)
    img = cv2.resize(img, (32,32))/255
    return img
```

Pre processing the Test data images and evaluating the model of test data

```python
test_imgs = []
for i in range(len(test_list)):
    test_imgs.append(np.reshape(get_test_image(test_list[i]), (32,32,3)))
test_imgs = np.array(test_imgs)
pred = model.predict(test_imgs)
```

Obtaining our results in the form of csv file and displaying our csv file with 'id' indicating the particular image and 'has cactus' indicating the variety of cactus 

```python
pred = [i for i in pred]

res_dict = {'id': test_list, 'has_cactus' : pred}
res_df = pd.DataFrame(res_dict)
res_df.to_csv('result.csv', index=False)
with open('result.csv', 'r') as csvFile:
    reader = csv.reader(csvFile)
    for row in reader:
        print(row)

csvFile.close()
```
